## Outlier

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv(r'C:\Users\ANIQ FARICHATUS\Downloads\archive (5)\googleplaystore.csv')

# Mengonversi kolom numerik yang relevan jika belum
data['Rating'] = pd.to_numeric(data['Rating'], errors='coerce')
data['Reviews'] = pd.to_numeric(data['Reviews'], errors='coerce')
data['Size'] = pd.to_numeric(data['Size'].str.replace('M', '').str.replace('k', ''), errors='coerce')
data['Installs'] = pd.to_numeric(data['Installs'].str.replace('[+,]', '', regex=True), errors='coerce')

# Menghilangkan baris dengan missing values (opsional, jika tidak ingin memproses missing data)
data_cleaned = data.dropna(subset=['Rating', 'Reviews', 'Size', 'Installs'])

# Menentukan kolom yang akan dianalisis untuk outlier
columns_to_analyze = ['Rating', 'Reviews', 'Size', 'Installs']

# Mencari outlier menggunakan metode IQR
def find_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1 
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Menemukan dan menampilkan outlier untuk masing-masing kolom yang dianalisis
for column in columns_to_analyze:
    print(f"Outlier di kolom {column}:")
    outliers = find_outliers(data_cleaned, column)
    print(outliers[[column]])
    print("\n")

# Visualisasi Boxplot untuk mendeteksi outlier
plt.figure(figsize=(12, 8))
for i, column in enumerate(columns_to_analyze, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x=data_cleaned[column])
    plt.title(f'Boxplot of {column}')
    
plt.tight_layout()
plt.show()
```

## Mengatasi Outlier

```python
import pandas as pd

# Menghapus outlier menggunakan metode IQR
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Menghapus outlier
    df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_filtered

# Menghapus outlier dari semua kolom yang dianalisis
for column in columns_to_analyze:
    data_cleaned = remove_outliers(data_cleaned, column)

# Menyimpan data tanpa outlier ke CSV baru
data_cleaned.to_csv('googleplaystore_cleaned.csv', index=False)

print(data_cleaned.head(10))
```

## Kategorisasi Variabel

```python
import pandas as pd

# Baca dataset
data = pd.read_csv(r'C:\Users\ANIQ FARICHATUS\Downloads\archive (5)\googleplaystore.csv')

# 1. Kategorisasi Rating (Low: < 2.5, Medium: 2.5 - 4, High: > 4)
def categorize_rating(rating):
    if rating < 2.5:
        return 'Low'
    elif 2.5 <= rating <= 4.0:
        return 'Medium'
    else:
        return 'High'

data_cleaned['Rating_Category'] = data_cleaned['Rating'].apply(categorize_rating)

# 2. Kategorisasi Reviews (Few: < 1,000, Moderate: 1,000 - 10,000, Many: > 10,000)
def categorize_reviews(reviews):
    if reviews < 1000:
        return 'Few'
    elif 1000 <= reviews <= 10000:
        return 'Moderate'
    else:
        return 'Many'

data_cleaned['Reviews_Category'] = data_cleaned['Reviews'].apply(categorize_reviews)

# 3. Kategorisasi Size (Small: < 10 MB, Medium: 10 - 30 MB, Large: > 30 MB)
def categorize_size(size):
    if size < 10:
        return 'Small'
    elif 10 <= size <= 30:
        return 'Medium'
    else:
        return 'Large'

data_cleaned['Size_Category'] = data_cleaned['Size'].apply(categorize_size)

# 4. Kategorisasi Installs (Low: < 100,000, Medium: 100,000 - 1,000,000, High: > 1,000,000)
def categorize_installs(installs):
    if installs < 100000:
        return 'Low'
    elif 100000 <= installs <= 1000000:
        return 'Medium'
    else:
        return 'High'

data_cleaned['Installs_Category'] = data_cleaned['Installs'].apply(categorize_installs)

# 5. Kategorisasi Price (Free: 0, Paid: > 0)
def categorize_price(price):
    if price == 0:
        return 'Free'
    else:
        return 'Paid'

data_cleaned['Price_Category'] = data_cleaned['Price'].apply(categorize_price)

# Menampilkan dataset dengan kategori baru
print(data_cleaned[['Rating', 'Rating_Category', 'Reviews', 'Reviews_Category', 'Size', 'Size_Category', 'Installs', 'Installs_Category', 'Price', 'Price_Category']].head())
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

data = pd.read_csv(r'C:\Users\ANIQ FARICHATUS\Downloads\archive (5)\googleplaystore.csv')

# Mendefinsikan kolom numerik
numeric_columns = ['Rating', 'Reviews', 'Size']

# Menghapus rows dengan NaN values pada kolom numerik
data = data.dropna(subset=numeric_columns)

# Convert 'Reviews' dan'Size' ke numerik
data['Reviews'] = pd.to_numeric(data['Reviews'], errors='coerce')
data['Size'] = pd.to_numeric(data['Size'].str.replace('M', '').str.replace('k', '').replace('Varies with device', np.nan), errors='coerce')
data = data.dropna(subset=numeric_columns)

# Visualisasi distribusi sebelum scaling
plt.figure(figsize=(12, 8))
for i, column in enumerate(numeric_columns, 1):
    plt.subplot(2, 2, i)
    sns.histplot(data[column], kde=True)
    plt.title(f'{column} - Before Scaling')
plt.tight_layout()
plt.show()

# 2. Transformasi data dengan StandardScaler
scaler = StandardScaler()
data_scaled = data.copy()
data_scaled[numeric_columns] = scaler.fit_transform(data[numeric_columns])

# 3. Visualisasi distribusi setelah scaling
plt.figure(figsize=(12, 8))
for i, column in enumerate(numeric_columns, 1):
    plt.subplot(2, 2, i)
    sns.histplot(data_scaled[column], kde=True)
    plt.title(f'{column} - After Scaling')
plt.tight_layout()
plt.show()
```
